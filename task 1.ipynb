#duplication


```python
data.duplicated().sum()  
```

```python
data = data.drop_duplicates()
```

```python
data['income'].unique()  
```

#column manage


```python
data = data.drop(columns=["education"])
```

```python
data.head(2)
```

```python
data = data.rename(columns={"id": "customer_id"})
data.head(2)
```

```python
data = data.reindex(sorted(data.columns), axis=1)
```

```python
data.head(2)
```

#missing values


```python
data.isna().sum()
```

#data type correction


```python
data['gender'] = pd.to_numeric(data['gender'].str.replace(',', ''), errors='coerce')
data.head(2)
```

#standardization


```python
data['loyalty_status'] = data['loyalty_status'].str.lower().str.strip()
data.head(2)
```

```python
data['region'] = data['region'].replace({
    'north': 'NORTH',
    'North': 'NORTH',
    'south': 'SOUTH',
    'South': 'SOUTH'
})
```
